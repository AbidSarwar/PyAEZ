## Importing Libraries

```python
'''import supporting libraries'''

import numpy as np
import matplotlib.pyplot as plt
import imageio

import sys
sys.path.append('./code/') # add path to source code directory
```

## Reading Data

```python
min_temp = np.load('./sample_data/input/climate/min_temp.npy') # Celcius
max_temp = np.load('./sample_data/input/climate/max_temp.npy') # Celcius
precipitation = np.load('./sample_data/input/climate/precipitation.npy') # mm / day
short_rad = np.load('./sample_data/input/climate/short_rad.npy') # W/m^2
wind_speed = np.load('./sample_data/input/climate/wind_speed.npy') # m/s
rel_humidity = np.load('./sample_data/input/climate/relative_humidity.npy') # Fraction, and assumed as a constant 0.70

admin_mask = imageio.imread('./sample_data/input/LAO_Admin.tif');
srtm_elevation = imageio.imread('./sample_data/input/SRTM_Elevation.tif'); # m
```

```python
import UtilitiesCalc
obj_utilities = UtilitiesCalc.UtilitiesCalc()

im_height = admin_mask.shape[0]
im_width = admin_mask.shape[1]

min_temp_d = np.zeros((im_height,im_width,365))
max_temp_d = np.zeros((im_height,im_width,365))
precipitation_d = np.zeros((im_height,im_width,365))
short_rad_d = np.zeros((im_height,im_width,365))
wind_speed_d = np.zeros((im_height,im_width,365))
rel_humidity_d = np.zeros((im_height,im_width,365))

for i_row in range(im_height):
    for i_col in range(im_width):
        min_temp_d[i_row,i_col,:] = obj_utilities.interpMonthlyToDaily(min_temp[i_row,i_col,:], 1, 365, False)
        max_temp_d[i_row,i_col,:] = obj_utilities.interpMonthlyToDaily(max_temp[i_row,i_col,:], 1, 365, False)
        precipitation_d[i_row,i_col,:] = obj_utilities.interpMonthlyToDaily(precipitation[i_row,i_col,:], 1, 365, True)
        short_rad_d[i_row,i_col,:] = obj_utilities.interpMonthlyToDaily(short_rad[i_row,i_col,:], 1, 365, True)
        wind_speed_d[i_row,i_col,:] = obj_utilities.interpMonthlyToDaily(wind_speed[i_row,i_col,:], 1, 365, True)
        rel_humidity_d[i_row,i_col,:] = obj_utilities.interpMonthlyToDaily(rel_humidity[i_row,i_col,:], 1, 365, True)
```

## Climate Regime Calculations

```python
'''importing of the ClimateRegime library. And passing climate data'''

import ClimateRegime

clim_reg = ClimateRegime.ClimateRegime()

# latitude limits of the study area
lat_min = 13.90
lat_max = 22.51

clim_reg.setStudyAreaMask(admin_mask, 0)
clim_reg.setLocationTerrainData(lat_min, lat_max, srtm_elevation)
clim_reg.setDailyClimateData(min_temp_d, max_temp_d, precipitation_d, short_rad_d, wind_speed_d, rel_humidity_d)
```

```python
'''importing of the UtilitiesCalc library for saving results'''

import UtilitiesCalc
obj_utilities = UtilitiesCalc.UtilitiesCalc()
```
